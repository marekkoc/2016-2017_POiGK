#### 3. Visualization Toolkit (VTK)

Biblioteka [VTK](http://www.vtk.org) jest zbiorem darmowych klas otwartego oprogramowania do tworzenia trójwymiarowej grafiki komputerowej, wizualizacji, posiada również możliwość przetwarzania obrazów. VTK jest napisane w języku C++ i jest dostępne dla głównych systemów operacyjnych: Windows, Linux, Unix i MacOS.

Proces generowania obrazów 2D lub 3D z wykorzystaniem komputera nazywany jest **renderowaniem**. Dane są zamieniane na obiekty graficzne, a następnie są wyświetlane na ekranie komputera. W prcesie tym odpwiednio modelowane są odbicia i załamania światła od powierzchni wyświetlanego obiektu, cienie, a nawet wpływ atmosfery (np. mgła).

W procesie renderowania możemy wyróżnić trzy podstawowe elementy: 
1. źródła światła,
1. wyświetlane obiekty (nazywane *aktorami*),
1. kamerę.

Najprostszym źródłem światła jest takie które umieszczono bardzo daleko od oświetlanej sceny (w *nieskończoności*), dzięki temu emitowane we wszystikich kierunkach promienie są równoległe względem siebie.

Biorąc pod uwagę ustawienie kamery w scenie renderowany obiekt może być widziany w różno sposób. W tym aspekcie możemy ustawić odpowiednie parametry kamery:
1. położenie,
1. orientację,
1. punkt ogniskowania kamery,
1. metodę rzutowania kamery (rzutowanie równoległe lub perspektywiczne),
1. współrzędne obcinania sceny przez kamerę.

Ważnym elementem sceny 3D jest obiekt który jest renderowany - *aktor*. W praktyce każdy taki obiekt jest reprezentowany przez zbiór podstawowych form geometrycznych (ang. *primitives*) takich jak: punkty, linie, wielokąty, krzywe itp.

Aby zbudować pełną scenę 3D i odpowienio zamodelowć obiekty (aktorów) w bibliotece VTK konieczne jest wykorzystanie siedmiu podsawowych elementów:
1. vtkRendererWindow - zarządza oknem pojawiającym się na ekranie monitora, wewnątrz takiego okna może znajdować się kilka rendererów.
1. vtkRenderer - koordynuje proces renderowania włączając w to obliczenia światła, kamery i wszystkich aktorów.
1. vtkLight - definiuje parametry źródła światła (lub kilku źródeł).
1. vtkCamera - definiuje kierunek patrzenia na scenę, punkt ogniskowania oraz inne parametry.
1. vtkActor - reprezentuje obiekt renderowany w scenie 3D wraz z jego właściwościami (ang. *properties*) takimi jak: kolor, typ cieniowania, rodzaj powierzchni, itp.. Aktor jest usytuowany w konkretnym miejscu sceny 3D - ma określone współrzędne w układzie kartezjańskim. 
1. vtkProperty - określa wygląd właściwości wyglądu aktora takie jak: kolor, przeźroczystość, reakcje na światło (np. stopień odbijania światła) oraz sposób reprezentacji za pomocą pełnych powierzchni, siatka lub punktów.
1. vtkMapper - zamienia geometrię która opisuje aktora na zbiór wielokątów łatwych do renderowania przez kartę graficzną. Kilku aktorów może być przypisanych do tego sameg mappera.

Biblioteka VTK jest tak zaprojektowana, że większość parametrów i ustawiwń związanych z kamerą, światłem i właściwościami aktorów jest dobierana w sposób automatyczny, a zarazem dosyć optymalny. Istnieje jednak możliwość modyfikacji wszystkich ustawień przez programistę.

Dokumentacja klas zawartych w VTK znajduje się na stronie domowej projektu [[1](http://www.vtk.org/documentation/),[2](http://www.vtk.org/doc/release/6.3/html/),[3](http://www.vtk.org/doc/release/6.3/html/annotated.html),[4](http://www.vtk.org/doc/release/6.3/html/classes.html)].
